```python
# Load the data into numpy matrices, skipping the header row
male = np.genfromtxt(male_file, delimiter=",", skip_header=1)
female = np.genfromtxt(female_file, delimiter=",", skip_header=1)
```

```python
# Calculate Body Mass Index (BMI) and add it as the 8th column to the female matrix
# BMI = weight (kg) / (height (m)^2)
female_heights_m = female[:, 1] / 100  # Convert height from cm to m
female_bmis = female[:, 0] / (female_heights_m ** 2)  # Calculate BMI
female = np.hstack((female, female_bmis.reshape(-1, 1)))  # Add BMI as the 8th column
```

```python
# Create a standardized version of the female dataset (z-scores)
female_means = female.mean(axis=0)
female_stds = female.std(axis=0)
zfemale = (female - female_means) / female_stds
```

```python
# Extract specific columns for scatterplot matrix (height, weight, waist circumference, hip circumference, BMI)
columns_indices = [1, 0, 6, 5, 7]  # Indices for height, weight, waist, hip, BMI
zfemale_selected = zfemale[:, columns_indices]
zfemale_labels = ["Height", "Weight", "Waist Circumference", "Hip Circumference", "BMI"]
```

```python
# Calculate waist-to-height ratio and waist-to-hip ratio for male and female participants
male_waist_to_height = male[:, 6] / male[:, 1]  # Waist circumference / Height
male_waist_to_hip = male[:, 6] / male[:, 5]     # Waist circumference / Hip circumference
male = np.hstack((male, male_waist_to_height.reshape(-1, 1), male_waist_to_hip.reshape(-1, 1)))

female_waist_to_height = female[:, 6] / female[:, 1]  # Waist circumference / Height
female_waist_to_hip = female[:, 6] / female[:, 5]     # Waist circumference / Hip circumference
female = np.hstack((female, female_waist_to_height.reshape(-1, 1), female_waist_to_hip.reshape(-1, 1)))
```

```python
# Create a standardized version of the female dataset (z-scores)
female_means = female.mean(axis=0)
female_stds = female.std(axis=0)
zfemale = (female - female_means) / female_stds

# Extract specific columns for scatterplot matrix (height, weight, waist circumference, hip circumference, BMI)
columns_indices = [1, 0, 6, 5, 7]  # Indices for height, weight, waist, hip, BMI
zfemale_selected = zfemale[:, columns_indices]
zfemale_labels = ["Height", "Weight", "Waist Circumference", "Hip Circumference", "BMI"]
```

```python
# Compute numerical aggregates
def describe_weights(data, label):
    mean = np.mean(data)
    median = np.median(data)
    std_dev = np.std(data)
    skewness = (3 * (mean - median)) / std_dev if std_dev != 0 else 0
    print(f"{label} Weights:")
    print(f"  Mean: {mean:.2f}")
    print(f"  Median: {median:.2f}")
    print(f"  Standard Deviation: {std_dev:.2f}")
    print(f"  Skewness: {skewness:.2f}")
    print()

describe_weights(female_weights, "Female")
describe_weights(male_weights, "Male")
```


**In your own words, list some advantages and disadvantages of BMI, waist-to-height ratio, and waist-to-hip ratio**
---

### **Body Mass Index (BMI)**  
#### **Advantages**:
1. **Ease of Calculation**: BMI is simple to calculate using weight and height.
2. **Widely Recognized**: It is a standard measure used globally for assessing obesity and underweight.
3. **Population-Level Analysis**: Useful for large-scale studies and public health assessments.
4. **Correlation with Health Risks**: BMI is linked to risks of conditions like diabetes, heart disease, and hypertension.

#### **Disadvantages**:
1. **Ignores Body Composition**: BMI cannot differentiate between muscle and fat, leading to misclassification of muscular individuals as overweight or obese.
2. **No Fat Distribution Insight**: It doesnt indicate where fat is distributed on the body, which is crucial for health risks.
3. **Age and Gender Variations**: It doesnt account for differences in body composition due to age or gender.
4. **Inaccurate for Some Groups**: Less effective for athletes, older adults, and children.

---

### **Waist-to-Height Ratio (WHtR)**  
#### **Advantages**:
1. **Focus on Central Fat**: WHtR measures abdominal fat, which is more strongly linked to health risks like cardiovascular disease.
2. **Applicable Across Groups**: More reliable than BMI across different ages, genders, and ethnicities.
3. **Quick and Simple**: Easy to measure with minimal equipment (a tape measure).
4. **Predictive Power**: Studies suggest WHtR is a better predictor of health risks than BMI.

#### **Disadvantages**:
1. **Less Standardized**: WHtR is not as widely used in clinical settings as BMI.
2. **Limited Public Awareness**: Most people are less familiar with interpreting WHtR compared to BMI.
3. **Requires Accurate Measurements**: Errors in measuring waist circumference or height can reduce reliability.

---

### **Waist-to-Hip Ratio (WHR)**  
#### **Advantages**:
1. **Measures Fat Distribution**: WHR captures fat stored around the waist relative to the hips, highlighting central obesity risks.
2. **Linked to Disease Risk**: High WHR is strongly associated with cardiovascular disease and metabolic syndrome.
3. **Better Predictor for Some Groups**: Particularly effective for older adults, where fat distribution is more critical than overall body weight.

#### **Disadvantages**:
1. **Measurement Challenges**: Accurately measuring the waist and hips can be difficult and prone to error.
2. **Less Reliable for Certain Body Types**: WHR may not work well for individuals with uniform body shapes or those with very low muscle mass.
3. **Cultural and Gender Variations**: Fat distribution patterns vary across populations and genders, complicating universal thresholds.
4. **Less Familiar Metric**: Like WHtR, its not as commonly understood or applied as BMI.

---

### **Summary**:  
- **BMI** is widely recognized and easy to calculate but lacks precision in assessing body composition and fat distribution.  
- **WHtR** emphasizes central obesity and provides a better risk assessment for various health conditions.  
- **WHR** focuses on fat distribution and is particularly effective in identifying central obesity risks but can be challenging to measure accurately.  

Each metric has its strengths and limitations, and their utility often depends on the context, population, and specific health goals.

This part of the code focuses on standardizing measurements, calculating BMI, and identifying individuals with the lowest and highest BMIs. Here's a step-by-step explanation:

**Defining Functions**:
standardize(data)
: This function takes a dataset (data) and standardizes it using z-score normalization. It centers the data around zero by subtracting the mean and scales it by dividing by the standard deviation.

**load_data(url)**: This function loads data from a given URL into a NumPy array. It assumes the data is comma-separated and skips the first row (header).

**def load_data(url)**: This function calculates the Body Mass Index (BMI) for each individual in the dataset. It assumes the first column (data[:, 0]) represents weight in kilograms and the second column (
data[:, 1]) represents height in centimeters. It converts height to meters and then applies the BMI formula: weight (kg) / (height (m)^2).
 You can switch to the female dataset by replacing male_data_url and female_data_url.

**bmi** = calculate_bmi(data):This line calculates the BMI for each individual in the loaded dataset using the
calculate_bmi function.

**standardized_data**:This line standardizes the first 7 columns of the dataset using the standarize function. This likely represents body measurements like weight, height, arm length, leg length, etc.

Identifying Individuals with Lowest and Highest BMI:

**lowest_bmi_indices**:This line finds the indices of the 5 individuals with the lowest BMI values using np.argsort, which sorts the BMI array and returns the indices of the sorted elements.
**highest_bmi_indices**:This line finds the indices of the 5 individuals with the highest BMI values.

```python
# Standardize measurements: (value - mean) / std
def standardize(data):
    return (data - np.mean(data, axis=0)) / np.std(data, axis=0)

# Load the necessary data (replace with your actual data loading)
def load_data(url):
    return np.genfromtxt(url, delimiter=",", skip_header=1)

# Calculate BMI
def calculate_bmi(data):
    # Assuming 'BMXWT' is weight in kg and 'BMXHT' is height in cm
    heights_m = data[:, 1] / 100  # Convert height from cm to m
    bmis = data[:, 0] / (heights_m ** 2)  # Calculate BMI
    return bmis

import numpy as np

# Choose either male_data_url or female_data_url
data = load_data(male_data_url) # or data = load_data(female_data_url)

bmi = calculate_bmi(data) # Calculate BMI outside the standardize function
standardized_data = standardize(data[:, :7]) # Standardize only the relevant columns

# Indices for 5 lowest and 5 highest BMI
lowest_bmi_indices = np.argsort(bmi)[:5]
highest_bmi_indices = np.argsort(bmi)[-5:]

lowest_bmi_data = standardized_data[lowest_bmi_indices]
highest_bmi_data = standardized_data[highest_bmi_indices]

print("Standardized Measurements for 5 Persons with Lowest BMI:")
print(lowest_bmi_data)

print("\nStandardized Measurements for 5 Persons with Highest BMI:")
print(highest_bmi_data)
```
